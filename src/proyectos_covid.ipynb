```python
#%matploplib inline 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filename = '../input/fondos/BDH_Proyectos.csv'
dataProyectos = pd.read_csv(filename,sep=';',encoding = "ISO-8859-1")
```

```python
#dataProyectos.head()
columnas = list(dataProyectos.columns)
```

```python
datawrite = dataProyectos[dataProyectos.NOMBRE_CONCURSO == "CONCURSO PARA LA ASIGNACION RAPIDA DE RECURSOS PARA PROYECTOS DE INVESTIGACION SOBRE EL CORONAVIRUS (COVID-19)"]
datawrite.to_excel('../output/resultados_concurso_covid.xlsx',encoding = "ISO-8859-1")
```

```python
data = datawrite[['MACROZONA MINCIENCIA (segUn ax)','REGION_EJECUCION','MONTO ADJUDICADO (MILES $)']]

resultados = data.groupby(['MACROZONA MINCIENCIA (segUn ax)','REGION_EJECUCION'])['REGION_EJECUCION'].count()

resultados
```

```python

data = datawrite[['REGION_EJECUCION','SEXO']]

resultados = data.groupby(['SEXO'])['SEXO'].count()

resultados

plt.figure(figsize=(5,5))
_ = plt.plot(resultados.index,resultados.values)



```

```python
data = datawrite[['INSTITUCION_PRINCIPAL','MONTO ADJUDICADO (MILES $)']]

resultados = data.groupby(['INSTITUCION_PRINCIPAL','MONTO ADJUDICADO (MILES $)']).sum()

resultados

resultados.to_excel('../output/resultados_concurso_covid.xlsx',encoding = "ISO-8859-1")

```

```python
data = datawrite[['REGION_EJECUCION','TIPO_BENEFICIARIO']]

resultados = data.groupby('TIPO_BENEFICIARIO')['TIPO_BENEFICIARIO'].count()


plt.figure(figsize=(5,5))
_ = plt.plot(resultados.index,resultados.values)


resultados
```
