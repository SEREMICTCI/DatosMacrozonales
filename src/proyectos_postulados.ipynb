```python
data = dataProyectos[dataProyectos.TIPO_DE_INSTITUCION =="UNIVERSIDAD"].groupby('ESTADO_RESOLUCION_CONCURSO').count()

dataResolucion = plt.plot(data.index,data.values)


_ = plt.show()
```

```python
dataRegionTarapaca = dataProyectos[dataProyectos.REGION_DE_EJECUCION =="01. TARAPACA"].groupby('ESTADO_RESOLUCION_CONCURSO').count()
dataRegionAntofagasta = dataProyectos[dataProyectos.REGION_DE_EJECUCION =="02. ANTOFAGASTA"].groupby('ESTADO_RESOLUCION_CONCURSO').count()
dataRegionAtacama = dataProyectos[dataProyectos.REGION_DE_EJECUCION =="03. ATACAMA"].groupby('ESTADO_RESOLUCION_CONCURSO').count()
dataRegionArica = dataProyectos[dataProyectos.REGION_DE_EJECUCION =="15. ARICA Y PARINACOTA"].groupby('ESTADO_RESOLUCION_CONCURSO').count()

plt.plot(dataRegionTarapaca.index,dataRegionTarapaca.values) #, label = 'Tarapaca')
plt.plot(dataRegionAntofagasta.index,dataRegionAntofagasta.values) #, label = 'Antofagasta')
plt.plot(dataRegionAtacama.index,dataRegionAtacama.values) #, label = 'Atacama')
plt.plot(dataRegionArica.index,dataRegionArica.values) #, label = 'Arica')

plt.legend(['Tarapaca','Antofagasta','Atacama','Arica'])
#plt.legend()
_= plt.show()
```
